# Data Visualization Using `plotnine`

The Palmer Penguins dataset had a variety of column types with which to explore some nice visualizations. The `mtcars` dataset is another popular dataset for doing some simple data work, but does not contain the same types of variables.

Run the following code to load the `mtcars` dataset and explore the observations and variables contained within. To learn more about this dataset check out [this site](https://stat.ethz.ch/R-manual/R-devel/library/datasets/html/mtcars.html)

```python
import statsmodels.api as sm
import pandas as pd

mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
df = pd.DataFrame(mtcars)
```

The statement "the `mtcars` dataset does not contain the same types of variables as the penguins dataset" is a little true and a little false. There are no variables that contain text values, BUT there are variables that it makes sense to consider categorical variables. In most situations, it makes sense to treat the values of categorical variables as text values instead of numeric values.

Use the `astype()` method to convert the categorical variables of the `mtcars` (df) dataset to have text values in the code chunk below. We've done one for you!

```python
df["am"] = df["am"].astype(str)

# Convert the other variables below
df["am"] = df["am"].astype(str)
df["cyl"] = df["cyl"].astype(str)
df["gear"] = df["gear"].astype(str)
df["carb"] = df["carb"].astype(str)
```

If you print out the dataset with the changed values, no visible difference will be seen; However, the variables that were transformed will now all be strings rather than floats, and this will change the functions that can be utilized. You can tell the variables became strings as they can now utilize functions like .cat(), but can no longer use functions like sum() and average(). Plotnine makes the variables discrete rather than continuous.

```python
# Create plot for (1) here
import plotnine as p9
(p9.ggplot(data=df, mapping=p9.aes(x='am', y='mpg')) + p9.geom_boxplot(alpha=0.))
```

```python
# Create plot for (1) here continued
df["am"] = df["am"].astype(float)
df["vs"] = df["vs"].astype(float)
(p9.ggplot(data=df, mapping=p9.aes(x='am', y='mpg')) + p9.geom_boxplot(alpha=0.))
```

# Create plot for (1) here continued
The first boxplot, where two boxplots are next to each other represents the transformation of the "am" variable into a string type. By doing so, it became discrete, and the two separate box plots allow us to compare the distributions of mpg for automatic versus manual transmissions. The singular boxplot that uses the float form of "am" treats the variable as continuous and places it on a continuous x axis from 0-1, the range of values. This makes the axis show a range of values rather than groupings, as the string/discrete method does, and it does not allow us to compare groups.

```python
# Create plot for (2) here
(p9.ggplot(df, p9.aes(x="hp", fill="cyl")) + p9.geom_histogram(alpha=0.4, position="identity", bins=15) )
#position = identity makes the boxes overlap
#bins = 15, chooses the number of equal width boxes to use
```

```python
# Create plot for (2) here continued
df["cyl"] = df["cyl"].astype(float)
(p9.ggplot(df, p9.aes(x="hp", fill="cyl")) + p9.geom_histogram(alpha=0.4, position="identity", bins=15) )
```

# Create plot for (2) here continued
In the first plot where the "cyl" variable is treated as a string, plotnine treats "cyl" as a categorical variable. Using fill, separate boxes of the histogram are created for each value of # of cylinders in a vehicle and each group is assigned a separate color. Using the position = "identity" part of the function allows us to overlap the boxes and compare the distributions of hp across different #'s of cylinders.

The second plot shows the difference when "cyl" is treated as a continuous numeric variable by plotnine when "cyl" is a float type variable. Fill no longer creates different groups since it is continuous, meaning no comparison can be made between distinct groups of cylinder #s. This creates a single histogram with a constant grey gradient losing any informational value gained previously.

Some of these variables, like the `cyl` variable, have numeric values that actually make sense as numbers (i.e. the number of cylinders in the engine). However, it doesn't make the most sense to "do math" with this type of variable (e.g. take averages and such) because there are so few different value this can take on AND they're an explicit choice made by the car manufacturer. So, it makes more sense to treat the `cyl` variable as a categorical variable despite it having numeric values.

#Create Plot One: Scatter Plot

```python
# Create plot with first new geometry here
from plotnine import *

(
    ggplot(df, aes("wt", "qsec")) #sets the x and y axis
    + geom_point(aes(fill="gear"), size=5) # sets the grouping for scatter plot
    + geom_point(aes(shape="gear"))
    + labs(title = "Weight(1000s of lbs) vs. 1/4 Mile Time by Number of Gears", x = "Weight(1000s of lbs)", y = "1/4 Mile Time(seconds)"
)
)

```

#Create Plot Two: Scatter Plot w/ Regression Line

```python
# Create plot with second new geometry here
from plotnine import * #all necessary libraris from plotnine

(
    ggplot(df, aes(x="hp", y="mpg", color="factor(cyl)"))
    + geom_point() # creates a scatterplot
    + geom_smooth(method="lm", se=True)  # regression lines with confidence interval shading for each line
    + labs(title="Horsepower vs. MPG by Cylinder Count", x="Horsepower",y="Miles per Gallon", color="Cylinders" #creates title for graph, x and y axis titles, and title for legend of colors
    )
)
```

#Create Plot Three: Grouped Bar Plot
