## Part Two: Decoding the secret message.


```python
top_secret = pd.read_csv("https://www.dropbox.com/s/k72h1zewk4gtqep/PA_Secret_Code?dl=1", header=None)

top_secret = top_secret[0].to_numpy()

top_secret = top_secret + 14
top_secret = top_secret * 18
top_secret = top_secret - 257
top_secret = np.exp(top_secret)
top_secret = top_secret ** 2
top_secret
#makes all changes to top secret

#checkpoint
x = 17

total = sum(top_secret < x)
total
#checks for 352 values under 17
```

```python
top_secret = top_secret.reshape(-1,5,order = 'F')
top_secret
#reshapes the array to 5 columns, sets the rows to -1 which allows numpy/python to automatically detemine the # of rows, order = 'F' makes it column wise
```

```python
evens = np.arange(1,383)
evens = evens[evens % 2 == 0]
len(evens)

#makes array with all even values 1-382
```

```python
top_secret[:,0] = top_secret[:,0] - evens
#Subtract the "evens" vector from the first column of your secret message matrix.

top_secret[18:25,2] = top_secret[18:25,2] - 100
#Subtract 100 from all numbers in the 18-24th rows of the 3rd column.

top_secret[:,3:5] = top_secret[:,3:5]*2
#Multiply all numbers in the 4th and 5th column by 2.

top_secret = top_secret.reshape(-1,order = 'F')
#reshapes the array to 1 columns, sets the rows to -1 which allows numpy/python to automatically detemine the # of rows, order = 'F' makes it column wise
```

```python
#checkpoint

#Check using If statement
checkactual = top_secret[500:].size
checkbelow = np.sum(top_secret[500:] < 100)
if checkbelow == checkactual:
  print("True")
  #prints true if equal

else:
  print("False")
  #prints false if not equal
#all values in indices 500 and below are below 100, compares total number of values above indice 500 to values below 100 in indices 500 and beyond


#Quicker Check using Numpy function .all
np.all(top_secret[500:] < 100)
```

```python
top_secret[37:465] = np.sqrt(top_secret[37:465])
#takes square root of all numbers in indices 37 to 464

top_secret = np.round(top_secret, 0)
#rounds all values to nearest whole number

top_secret = np.where(top_secret == 39, 20, top_secret)
#numpy command to replace all instances of the number 39 with 20

```

```python
#checkpoint

checksecret = np.sum(top_secret % 2 == 0)
checksecret

if checksecret == 344:
  print("True")
  #prints true if equal
else:
  print("False")
  #prints false if not equal

#checks if there are 344 even numbers
```
